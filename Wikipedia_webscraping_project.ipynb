## Step 1: Setting up

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0 Safari/537.36"
}
page = requests.get(url, headers=headers)

soup = BeautifulSoup(page.text, 'html.parser')
```

```python
#print(soup)
```

## Step 2: Getting the headers from the first table

```python
table = soup.find_all('table')[0] #since its a list we can use indexing
```

```python
titles = table.find_all('th') #using table instead of soup so we only pull from the first table. Soup will pull all the th from every table
```

```python
titles
```

```python
table_titles = [headers.text.strip() for headers in titles]

print(table_titles)
```

```python
df = pd.DataFrame(columns  = table_titles)
df
```

## Step 3: Getting the data from the td tags in the first table.

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:  #starting from position 1 because position 0 is empty list[] which messes up the appending of data into the headers
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    length = len(df) #checking the length of our data frame
    df.loc[length] = individual_row_data #appending each row of information into the next position in our data frame
```

```python
df
```

## Step 4: Exporting to csv

```python
df.to_csv(r'my-path\Companies.csv', index = False)
```
